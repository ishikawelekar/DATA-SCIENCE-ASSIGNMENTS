```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
x_train = train.drop(['maritalstatus','sex','native','race','relationship','education','Salary'],axis=1).values
y_train = train['Salary'].values                      
```

```python
x_test = test.drop(['maritalstatus','sex','native','race','relationship','education','Salary'],axis=1).values
y_test = test['Salary'].values
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

# Without Using TFIDF

## Multinomial Navie Bayes

```python
#On train dataset
classifier_mb = MB()
classifier_mb.fit(x_train,y_train)
train_pred_mb = classifier_mb.predict(x_train)
accuracy_train_mb = np.mean(train_pred_mb==y_train)
accuracy_train_mb
```

```python
#On test dataset
test_pred_mb = classifier_mb.predict(x_test)
accuracy_test_mb = np.mean(test_pred_mb==y_test)
accuracy_test_mb
```

```python
#On train dataset
classifier_gb = GB()
classifier_gb.fit(x_train,y_train)
train_pred_gb = classifier_gb.predict(x_train)
accuarcy_train_gb = np.mean(train_pred_gb==y_train)
accuarcy_train_gb
```

```python
#On test dataset
test_pred_gb = classifier_gb.predict(x_test)
accuracy_test_gb = np.mean(test_pred_gb==y_test)
accuracy_test_gb
```
