# ForestFires Dataset

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
forest=pd.read_csv("D://DATA SCIENCE//ASSIGNMENTS//SVM//forestfires.csv")
```

```python
forest
```

```python
forest.info()
```

```python
forest.describe()
```

```python
forest.isnull().sum()
```

```python
X= forest.iloc[:,0:30]
Y= forest.iloc[:,30]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X= X.apply(LabelEncoder().fit_transform)
```

```python
X
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.3, random_state=40)
```

```python
model_linear = SVC(kernel= "linear")
model_linear.fit(x_train,y_train)
preds_linear = model_linear.predict(x_test)
np.mean(preds_linear==y_test)
```

```python
model_poly = SVC(kernel="poly")
model_poly.fit(x_train,y_train)
preds_poly = model_poly.predict(x_test)
np.mean(preds_poly==y_test)
```

```python
model_rbf = SVC(kernel="rbf")
model_rbf.fit(x_train,y_train)
preds_rbf=model_rbf.predict(x_test)
np.mean(preds_rbf==y_test)
```

```python
model=SVC()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print(classification_report(pred,y_test))
```

```python
np.mean(pred==y_test)
```
