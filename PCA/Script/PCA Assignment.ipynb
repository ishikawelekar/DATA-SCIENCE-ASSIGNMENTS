# PCA 

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv("D:\DATA SCIENCE\ASSIGNMENTS\PCA\wine.csv")
wine
```

```python
wine.info()
```

```python
wine.describe()
```

```python
#Considering only numerical data
wine1=wine.iloc[:,1:]
wine1.head()
```

```python
#converting into numpy array
win=wine1.values
win
```

```python
#normalizing the numerical data
wine_normal=scale(win)
wine_normal
```

# PCA Implementation

```python
pca=PCA()
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
#PCA components matrix
pca.components_
```

```python
#the amount of variance that each PCA explains is
var=pca.explained_variance_ratio_
var
```

```python
#cumulative variance
var1=np.cumsum(np.round(var,decimals=4)*100)
```

```python
var1
```

```python
#variance plot for PCA components obtained
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
#plot between PCA1 and PCA2
x=pca_values[:,0:1]
y=pca_values[:,1:2]
z=pca_values[:,2:3]
plt.scatter(x,y)
```

```python
y
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3'])],axis=1)
final_df
```

```python
#visualization of pca
import seaborn as sns
fif=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df,x='pc1',y='pc2',hue='Type');
```

# Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#create dendrogram
plt.figure(figsize=(16,12))
dendrogram=sch.dendrogram(sch.linkage(wine_normal,method='complete'))
```

```python
#create clusters for complete
hc=AgglomerativeClustering(n_clusters=4, affinity='euclidean',linkage='complete')
```

```python
#save clusters for chart
y_hc=hc.fit_predict(wine_normal)
clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
clusters
```

```python
#addig clusters to dataset
wine2=wine.copy()
wine2['Clusters']=hc.labels_
```

```python
wine2
```

# KMeans

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
```

```python
X=np.random.uniform(0,1,177)
Y=np.random.uniform(0,1,177)
df_xy=pd.DataFrame(columns=["X","Y"])
df_xy.X=X
df_xy.Y=Y
df_xy.plot(x="X",y="Y",kind="scatter")
```

```python
model1=KMeans(n_clusters=4).fit(df_xy)
df_xy.plot(x="X",y="Y",c=model1.labels_,kind="scatter",s=10,cmap=plt.cm.coolwarm)
```

```python
#selecting 3 clustering from the normalize dataset
model=KMeans(n_clusters=3)
model.fit(wine_normal)
model.labels_
```

```python
md=pd.Series(model.labels_)
wine['clust']=md
```
